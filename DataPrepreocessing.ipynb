# Separate The Minor Class From The Major Class To Apply The Resampling Technique

```python
# oversampling or undersampling 
from sklearn.utils import resample

fraud_data = df[df['Class'] == 1 ]
nonfraud_data = df[df['Class'] == 0 ]

# oversampling minor--> fraud data 492 example

the_resampled_minor_class = resample(fraud_data, replace=True, random_state=42, n_samples= len(nonfraud_data))

the_resampled_minor_class
```

# balance The Dataset Using Balancing Techniques, Used Libraries: Sklearn.

```python
balanced_dataset = pd.concat([nonfraud_data,the_resampled_minor_class])

balanced_dataset = balanced_dataset.sample(frac=1).reset_index(drop = True)

balanced_dataset
```
